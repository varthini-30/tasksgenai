```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
tokenizer.fit_on_texts([data])
tokenizer.word_index
```

TOKENIZATION-pieces of text
if i want characters no need of []
if i want words use[]-item handling

```python
from typing import Sequence
input_sequence=[]
for sentence in data.split('\n'):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequence.append(tokenized_sentence[:i+1])
input_sequence
```

text data->sequence i+1 slicing nxt

```python
max_len=max([ len (x) for x in input_sequence])
max_len
```

converting to ml
[440, 4],need 1
[301, 441],need 1
[301, 441, 442],
pairing method -matchup
->padding=putting 0 (dummy)
 pre and post pdding
 maximum length of matrix


```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padd=pad_sequences(input_sequence,maxlen=max_len,padding="pre")
padd
```

```python
x=padd[:,:-1]
y=padd[:,-1]
y
```

```python
print(x.shape)
print(y.shape)
print(len(tokenizer.word_index))
```

as of now we are using  softmax->multiclassification


```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=1783)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
model=Sequential([
    Embedding(1783,100,input_length=16),
    SimpleRNN(150),
    Dense(1783,activation="softmax")
])
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(x,y,epochs=20)
```

```python
import numpy as np
import time
text="where"
for i in range(10):
  ss=tokenizer.texts_to_sequences([text])[0]
  dd=pad_sequences([ss],maxlen=max_len-1,padding="pre")
  pos=np.argmax(model.predict(dd))
  for word,index in tokenizer.word_index.items():
    if pos==index:
      text=text +" "+word
      print(text)
      time.sleep(1)
```
